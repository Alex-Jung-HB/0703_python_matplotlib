matplotlib.pyplot이란?
matplotlib.pyplot은 MATLAB과 유사한 인터페이스를 제공하는 그래프 그리기 도구입니다.

plt는 **matplotlib.pyplot 모듈의 별칭(alias)**이고 파이썬 데이터 시각화에서 가장 기본이 되는 도구입니다!

주요 함수들<br>
plt.plot() - 선 그래프<br>
plt.scatter() - 산점도<br>
plt.bar() - 막대 그래프<br>
plt.hist() - 히스토그램<br>
plt.show() - 그래프 화면에 표시<br>
plt.xlabel() - x축 라벨<br>
plt.ylabel() - y축 라벨<br>
plt.title() - 제목

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np  # numpy 추가 필요

def setup_basic_plot(step_num=1, action='forward'):
    print("===Basic plot example ===")
    print(f"Step: {step_num}, Action: {action}")

    # 랜덤 데이터 생성
    x = np.random.randn(50)   # 표준정규분포: 대부분 -3 ~ +3 사이
    y = np.random.randn(50)   # 표준정규분포: 대부분 -3 ~ +3 사이
    colors = np.random.rand(50)

    # 폰트 설정
    plt.rcParams['font.family'] = 'DejaVu Sans'
    plt.rcParams['font.size'] = 12

    # 플롯 생성
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, c=colors, alpha=0.7)  # 산점도 그리기: x, y = 좌표, c = 랜덤색상, alpha = 색상투명도
    plt.xlim(-4, 4)       # x축: -4부터 4까지
    plt.ylim(-4, 4)       # y축: -4부터 4까지
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Random Scatter Plot - {action}')
    plt.grid(True, alpha=0.5)
    plt.tight_layout()  # 그래프의 제목, 축 라벨, 범례 등이 겹치지 않도록 자동으로 여백 조정
    plt.show()

setup_basic_plot(5, "turn_left")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def calculate_goal_line(obstacles_y):
    """목표선 Y좌표 계산 함수"""
    if len(obstacles_y) == 0:
        goal_y = 0 + 3  # 장애물이 없으면 Y=0 기준
    else:
        max_obstacle_y = np.max(obstacles_y)  # 가장 높은 장애물
        goal_y = max_obstacle_y + 3  # 최대 Y좌표 + 3
    return goal_y

def draw_goal_line(obstacles, action="forward"):
    """목표선을 그리는 함수"""
    # 폰트 설정
    plt.rcParams['font.family'] = 'DejaVu Sans'
    plt.rcParams['font.size'] = 12

    # 캔버스 설정
    plt.figure(figsize=(8, 6))

    # 축 범위 설정 (수정됨)
    plt.xlim(-1, 8)       # x축: -1부터 8까지
    plt.ylim(-1, 12)      # y축: -1부터 12까지 (수정됨)

    plt.grid(True, alpha=0.3)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Goal Line Plot - {action}')

    # 장애물 데이터 분리
    x_coords = [point[0] for point in obstacles]
    y_coords = [point[1] for point in obstacles]

    # 장애물 표시
    plt.scatter(x_coords, y_coords, c='red', s=100, alpha=0.7, label='Obstacle')

    # 목표선 계산
    goal_y = calculate_goal_line(y_coords)

    # 목표선 그리기 (수정됨)
    plt.axhline(y=goal_y, color='lime', linewidth=3, alpha=0.7, label='Target Line')

    # 텍스트 표시
    plt.text(7, goal_y + 0.2, f"GOAL (y={goal_y})",
             color='green',
             fontsize=12,
             fontweight='bold')

    # 최대 장애물 찾기 및 표시
    if len(y_coords) > 0:
        max_y = max(y_coords)
        max_idx = y_coords.index(max_y)
        plt.scatter(x_coords[max_idx], y_coords[max_idx],
                   c='orange', s=150, marker='*',
                   label=f'Highest Obstacle (y={max_y})')
        print(f"Max_Y-Coodinate: {max_y}")
        print(f"Y-Coodinate of Taget Line: {goal_y}")

    plt.legend()
    plt.tight_layout()
    plt.show()

    # 실행 코드
if __name__ == "__main__":
    # 장애물 좌표 설정
    obstacles = [(3, 2), (3, 3), (2, 4), (4, 5)]

    # 함수 실행
    draw_goal_line(obstacles, "forward")
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np

def draw_obstacles(obstacles, action = "forward"):

    # 캔버스 설정
    fig, ax = plt.subplots(figsize=(8, 6))    # Figure = 전체 캔버스/종이, Axes = 실제 데이터가 그려지는 영역

    # 축 범위 설정
    ax.set_xlim(-1, 8)       # x축: -1부터 8까지
    ax.set_xlim(-1, 12)      # y축: -1부터 12까지

    ax.grid(True, alpha=0.3)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title(f'Goal Line Plot - {action}')    # set_title로 수정

    # 장애물 그리기(Rectangle 객체 생성)
    for i, (x, y) in enumerate(obstacles):
      width = 0.8        # width 변수 정의
      height = 0.8       # width 변수 정의

      rect = Rectangle(
          (x - width/2, y - height/2),    # 왼쪽 아래 모서리 좌표
          width, height,                  # 너비, 높이(중심 좌표에서 0.4씩 빼서 위치 조정)
          linewidth=2,                    # 테두리 두께
          edgecolor='red',                # 테두리 색상
          facecolor='none',               # 내부 채우기 색상
          alpha=0.7                       # 투명도
      )

    # 사각형을 축에 추가
      ax.add_patch(rect)

    # 텍스트 표시
      ax.text(x, y, f'{i+1}',          # 번호는 1부터 설정(프로그램에선 0부터 시작하므로 '+1')
          ha='center', va='center',  # 완전 중앙 정렬, ha = horizontal alignment (수평 정렬)
          fontsize=12, fontweight='bold',
          color='blue',              # 흰색 텍스트
          bbox = dict(boxstyle = "circle, pad = 0.1",
                      facecolor = "navy",
                      edgecolor = "white",
                      alpha = 0.9))    # 원형 배경 추가

plt.tight_layout()
plt.show()

def calculate_goal_line(obstacles_y):
    """목표선 Y좌표 계산 함수"""
    if len(obstacles_y) == 0:
        goal_y = 0 + 3  # 장애물이 없으면 Y=0 기준
    else:
        max_obstacle_y = np.max(obstacles_y)  # 장애물이 있는 경우 가장 높은 장애물을 찾아서 지정
        goal_y = max_obstacle_y + 3  # 최대 Y좌표 + 3
    return goal_y

def draw_goal_line(obstacles, action="forward"):
    """목표선을 포함한 완전한 그래프"""

    # 장애물 그리기
    draw_obstacles(obstacles, action)

    # 목표선 계산 및 추가
    fig, ax = plt.subplots(figsize=(8, 6))    # Figure = 전체 캔버스/종이, Axes = 실제 데이터가 그려지는 영역
    ax.set_xlim(-1, 8)       # x축: -1부터 8까지
    ax.set_xlim(-1, 12)      # y축: -1부터 12까지
    ax.grid(True, alpha = 0.3)
    ax.set_xlabel('X')
    ax.set_xlabel('Y')
    ax.set_title(f'Goal Line Plot - {action}')

    # 장애물 다시 그리기
    for i, (x, y) in enumerate(obstacles):
        width = 0.8        # width 변수 정의
        height = 0.8       # width 변수 정의

        rect = Rectangle(
            (x - width/2, y - height/2),    # 왼쪽 아래 모서리 좌표
            width, height,                  # 너비, 높이(중심 좌표에서 0.4씩 빼서 위치 조정)
            linewidth=2,                    # 테두리 두께
            edgecolor='red',                # 테두리 색상
            facecolor='none',               # 내부 채우기 색상
            alpha=0.7                       # 투명도
        )

    # 사각형을 축에 추가
        ax.add_patch(rect)

    # 텍스트 표시
        ax.text(x, y, f'{i+1}',             # 번호는 1부터 설정(프로그램에선 0부터 시작하므로 '+1')
            ha='center', va='center',       # 완전 중앙 정렬, ha = horizontal alignment (수평 정렬)
            fontsize=12, fontweight='bold',
            color='blue')                   # 파란색 텍스트

    # 목표선 계산 및 그리기
    y_coords = [point[1] for point in obstacles]
    goal_y = calculate_goal_line(y_coords)

    # 목표선 그리기
    ax.axhline(y=goal_y, color='lime', linewidth=3,         # 축에 수평선을 그리는 함수
               linestyle='--', alpha=0.8, label='목표선')

    # 목표 텍스트 표시
    ax.text(7, goal_y + 0.2, f"GOAL (y={goal_y})",
            color='green',
            fontsize=12,
            fontweight='bold',
            bbox=dict(boxstyle="round,pad=0.3",
                     facecolor="lightgreen", alpha=0.8))

    ax.legend()
    plt.tight_layout()
    plt.show()

    print(f"목표선 Y좌표: {goal_y}")

    # 실행 코드 (함수 정의 외부로 이동)
if __name__ == "__main__":
    # 장애물 좌표 설정
    obstacles = [(3, 2), (3, 3), (2, 4), (4, 5)]

    # 함수 실행
    print("=== 장애물만 그리기 ===")
    draw_obstacles(obstacles, "forward")

    print("=== 목표선 포함 그리기 ===")
    draw_goal_line(obstacles, "forward")
```
