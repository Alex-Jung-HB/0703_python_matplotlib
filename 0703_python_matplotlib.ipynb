matplotlib.pyplot이란?
matplotlib.pyplot은 MATLAB과 유사한 인터페이스를 제공하는 그래프 그리기 도구입니다.

plt는 **matplotlib.pyplot 모듈의 별칭(alias)**이고 파이썬 데이터 시각화에서 가장 기본이 되는 도구입니다!

주요 함수들<br>
plt.plot() - 선 그래프<br>
plt.scatter() - 산점도<br>
plt.bar() - 막대 그래프<br>
plt.hist() - 히스토그램<br>
plt.show() - 그래프 화면에 표시<br>
plt.xlabel() - x축 라벨<br>
plt.ylabel() - y축 라벨<br>
plt.title() - 제목

```python
import matplotlib
print(matplotlib.__version__)
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

예제 1: 기본 선 그래프

```python
import matplotlib.pyplot as plt
import numpy as np

def example1_basic_line():
    print("=== 예제 1: 기본 선 그래프 ===")
    x = np.array([0, 1, 2, 3, 4, 5])
    y = np.array([0, 1, 4, 9, 16, 25])

    plt.figure(figsize=(8, 6))
    plt.plot(x, y)
    plt.title('기본 선 그래프')
    plt.xlabel('X축')
    plt.ylabel('Y축')
    plt.show()

example1_basic_line()   # 위에서 정의한 함수를 호출해야 결과 나옴!
```

```python
import matplotlib.pyplot as plt
import numpy as np

print("=== 예제 1: 기본 선 그래프 ===")
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 4, 9, 16, 25])

plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title('기본 선 그래프')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

예제 2: 산점도 (Scatter Plot)

```python
plt.rc('font', family='NanumGothic')  # 한글 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

def example2_scatter():
    print("=== 예제 2: 산점도 ===")
    x = np.random.randn(50)
    y = np.random.randn(50)
    colors = np.random.rand(50)

    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, c=colors, alpha=0.7)
    plt.title('컬러 산점도')
    plt.xlabel('X값')
    plt.ylabel('Y값')
    plt.colorbar()
    plt.show()

example2_scatter()   # 위에서 정의한 함수를 호출해야 결과 나옴!
```
